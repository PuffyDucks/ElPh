```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##### This notebook is for visualizing the map of the electron-phonon coupling w.r.t phonon frequency #####

# Define constants
#kb = 1.38 * 1e-23 # Boltzmann constant (J/K)
#temp = 298 # Temperature, we set as room temperature (K)
thztocm_1 = 33.356 # Conversion factor from THz to cm-1

# Load Phonon frequency 
freq = np.loadtxt("frequencies.txt")

# Load electron-phonon coupling 
epc = np.load('ep_coupling.npz')

fig, axes = plt.subplots(3, 1, sharex=True, figsize=(6, 8))

for q in range(len(freq)):
    axes[0].bar(freq[q]*thztocm_1, epc['A'][q]**2 * 10000, width=0.5, color='red')
    axes[1].bar(freq[q]*thztocm_1, epc['B'][q]**2 * 10000, width=0.5, color='blue')
    axes[2].bar(freq[q]*thztocm_1, epc['C'][q]**2 * 10000, width=0.5, color='green')

axes[0].set_ylabel(r'$g_{ij}^2$ x 1$e^4$')
axes[0].set_title('Electron Phonon Coupling Parameter vs Frequency for Dimer A')
axes[0].set_yscale('log')
axes[0].set_ylim(1e-4,1e0)

axes[1].set_ylabel(r'$g_{ij}^2$ x 1$e^4$')
axes[1].set_title('Electron Phonon Coupling Parameter vs Frequency for Dimer B')
axes[1].set_yscale('log')
axes[1].set_ylim(1e-4,1e0)

axes[2].set_ylabel(r'$g_{ij}^2$ x 1$e^4$')
axes[2].set_title('Electron Phonon Coupling Parameter vs Frequency for Dimer C')
axes[2].set_yscale('log')
axes[2].set_ylim(1e-4,1e0)


axes[-1].set_xlim(-10, 3500) 
axes[-1].set_xlabel(r'$Energy$ (cm-1)')

plt.show()
```

```python
var = np.load('variance.npz')
fig, axes = plt.subplots(3, 1, sharex=True, figsize=(6, 8))

for q in range(len(freq)):

    axes[0].bar(freq[q]*thztocm_1, var['A'][q], width=0.5, color='red')
    axes[1].bar(freq[q]*thztocm_1, var['B'][q], width=0.5, color='blue')
    axes[2].bar(freq[q]*thztocm_1, var['C'][q], width=0.5, color='green')

axes[0].set_ylabel(r'$\sigma_{A}^2$')
axes[0].set_title('Variance vs Frequency for Dimer A')
axes[0].set_yscale('log')
axes[0].set_ylim(1e-8,1e-3)

axes[1].set_ylabel(r'$\sigma_{B}^2$')
axes[1].set_title('Variance vs Frequency for Dimer B')
axes[1].set_yscale('log')
axes[1].set_ylim(1e-8,1e-3)

axes[2].set_ylabel(r'$\sigma_{C}^2$')
axes[2].set_title('Variance vs Frequency for Dimer B')
axes[2].set_yscale('log')
axes[2].set_ylim(1e-8,1e-3)

axes[-1].set_xlim(-10, 3500) 
axes[-1].set_xlabel(r'$Energy$ (cm-1)')

plt.show()
```

```python
norm_var = (var['A'] + var['B'] + var['C'])**0.5
fig, ax = plt.subplots()

for q in range(len(freq)):

    ax.bar(freq[q]*thztocm_1, norm_var[q]*1e3, width=0.5, color='pink')

ax.set_ylabel(r'$\sigma$  (meV)')
ax.set_title('Transfer Integral Fluctuation vs Frequency for Anthracene')

ax.set_xlim(-10, 3500) 
ax.set_xlabel(r'$Energy$ (cm-1)')

plt.show()
```
